# User Data Gen

```python
import pandas as pd
from faker import Faker
import random

# Initialize Faker
fake = Faker()

# Sydney-specific neighborhoods and postcodes
sydney_neighborhoods = [
    {"neighborhood": "Sydney CBD", "postcode": "2000"},
    {"neighborhood": "Bondi Beach", "postcode": "2026"},
    {"neighborhood": "Manly", "postcode": "2095"},
    {"neighborhood": "Surry Hills", "postcode": "2010"},
    {"neighborhood": "Parramatta", "postcode": "2150"},
    {"neighborhood": "Chatswood", "postcode": "2067"},
    {"neighborhood": "North Sydney", "postcode": "2060"},
    {"neighborhood": "Paddington", "postcode": "2021"},
    {"neighborhood": "Newtown", "postcode": "2042"},
    {"neighborhood": "Darlinghurst", "postcode": "2010"}
]

# Generate User Data
def generate_data(users):
    user_data = []
    for user_id in range(1, users + 1):  # Generating 100 users
        location = random.choice(sydney_neighborhoods)
        user = {
            'user_id': user_id,
            'name': fake.name(),
            'email': fake.unique.email(),
            'phone': fake.phone_number(),
            'age': random.randint(20, 60),
            'gender': random.choice(['Male', 'Female']),
            'location': f"Sydney, {location['neighborhood']}, {location['postcode']}",
            'neighborhood': location['neighborhood'],
            'postcode': location['postcode'],
            'income_level': random.choice(['Low', 'Medium', 'High']),
            'family_status': random.choice(['Single', 'Married', 'With Kids']),
        }
        user_data.append(user)
    users_df = pd.DataFrame(user_data)
    return  users_df

# Save to CSV
# users_df.to_csv('users.csv', index=False)
# 
# # Display the first few rows
# print(users_df.head())
```

# Google Analytics Data Gen

```python

from datetime import datetime, timedelta

# Initialize Faker
fake = Faker()

# Load existing user and property data
users_df = pd.read_csv('users.csv')
properties_df = pd.read_csv('properties.csv')

# Event Types
event_types = ["Page View", "Property View", "Click", "Inquiry Submission"]

# Referral Sources
referral_sources = ["Direct", "Organic Search", "Social Media", "Email Campaign", "Referral"]

# Device Types
device_types = ["Desktop", "Mobile", "Tablet"]

# Browsers
browsers = ["Chrome", "Safari", "Firefox", "Edge"]

# Operating Systems
operating_systems = ["Windows", "macOS", "iOS", "Android"]

# Generate Enhanced Website Interaction Data
def generate_browser_data(users):
    analytics_data = []
    session_id = 1
    
    for _ in range(1, users + 1):  # Generating 500 sessions
        user_id = random.choice(users_df['user_id'])
        session_start = fake.date_time_between(start_date='-30d', end_date='now')
        session_duration = 0
        num_events = random.randint(1, 10)  # Number of events in a session
        device_type = random.choice(device_types)
        browser = random.choice(browsers)
        os = random.choice(operating_systems)
        referral_source = random.choice(referral_sources)
    
        for event_num in range(num_events):
            event_duration = random.randint(30, 600)  # Time spent on each event
            event_time = session_start + timedelta(seconds=session_duration)
            event_type = random.choice(event_types)
    
            # Link to a random property for Property View, Click, or Inquiry
            property_id = random.choice(properties_df['property_id']) if event_type in ["Property View", "Click", "Inquiry Submission"] else None
    
            event = {
                'event_id': session_id * 1000 + event_num,
                'session_id': session_id,
                'user_id': user_id,
                'property_id': property_id,
                'event_type': event_type,
                'event_timestamp': event_time,
                'session_duration': session_duration + event_duration,
                'referral_source': referral_source,
                'device_type': device_type,
                'browser': browser,
                'operating_system': os,
            }
            analytics_data.append(event)
    
            session_duration += event_duration  # Increment session duration
    
        session_id += 1
    analytics_df = pd.DataFrame(analytics_data)
    return analytics_df

# # Save to CSV
# analytics_df.to_csv('enhanced_website_interactions.csv', index=False)
# 
# # Display the first few rows
# print(analytics_df.head())
```

# Sales Interaction Data

```python
import pandas as pd
from faker import Faker
import random

# Initialize Faker for generating fake data
fake = Faker()

# Load existing data
properties_df = pd.read_csv('properties.csv')
users_df = pd.read_csv('users.csv')
agents_df = pd.read_csv('agents.csv')

# Define the number of transactions to generate
num_transactions = 100

# Generate Sales Data
sales_data = []
for transaction_id in range(1, num_transactions + 1):  # Generating 100 sales transactions
    # Randomly select a property, buyer, and agent
    property_row = properties_df.sample(n=1).iloc[0]
    buyer = users_df.sample(n=1).iloc[0]
    agent = agents_df.sample(n=1).iloc[0]
    
    # Generate transaction details
    sale_price = random.randint(property_row['price'] - 100000, property_row['price'] + 100000)  # Variance around the listed price
    commission_rate = random.uniform(0.02, 0.05)  # Random commission rate between 2% and 5%
    commission_earned = sale_price * commission_rate  # Calculate commission
    sale_status = random.choice(['Completed', 'Pending', 'Canceled'])  # Randomly assign a sale status
    
    # Create the transaction record
    transaction = {
        'Transaction ID': transaction_id,
        'Property ID': property_row['property_id'],  # Consistent with property data
        'Buyer ID': buyer['user_id'],  # Consistent with user data
        'Agent ID': agent['Agent ID'],  # Consistent with agent data
        'Sale Price': sale_price,
        'Sale Date': fake.date_between(start_date='-1y', end_date='today'),  # Sale date within the last year
        'Payment Method': random.choice(['Mortgage', 'Cash', 'Installments']),
        'Sale Status': sale_status,
        'Commission Earned': round(commission_earned, 2),  # Commission rounded to two decimal places
        'Property Tags': property_row['property_tags'],  # Tags from property data
        'Transaction Notes': fake.text(max_nb_chars=200) if sale_status != 'Completed' else "",  # Notes if not completed
    }
    sales_data.append(transaction)

# Convert to DataFrame
sales_df = pd.DataFrame(sales_data)

# Save the DataFrame to a CSV file
sales_df.to_csv('sales.csv', index=False)

# Display the first few rows of the DataFrame for verification
print(sales_df.head())
```

# CRM Data Gen

```python


# Initialize Faker for generating fake data
fake = Faker()

# Load existing data
users_df = pd.read_csv('users.csv')
agents_df = pd.read_csv('agents.csv')
properties_df = pd.read_csv('properties.csv')

# Define the number of interactions to generate
num_interactions = 200

# Interaction types and outcomes
interaction_types = ['Email', 'Phone Call', 'Meeting']
interaction_outcomes = ['Positive', 'Neutral', 'Negative']

# Generate CRM Interaction Data
crm_interactions = []
for interaction_id in range(1, num_interactions + 1):  # Generating 200 interactions
    # Randomly select a user, agent, and optionally a property
    user = users_df.sample(n=1).iloc[0]
    agent = agents_df.sample(n=1).iloc[0]
    property_row = properties_df.sample(n=1).iloc[0] if random.choice([True, False]) else None
    
    # Generate interaction details
    interaction_type = random.choice(interaction_types)
    interaction_outcome = random.choice(interaction_outcomes)
    
    # Create the interaction record
    interaction = {
        'Interaction ID': interaction_id,
        'User ID': user['user_id'],  # Consistent with user data
        'Agent ID': agent['Agent ID'],  # Consistent with agent data
        'Property ID': property_row['property_id'] if property_row is not None else None,  # Consistent with property data, optional
        'Interaction Type': interaction_type,
        'Interaction Date': fake.date_between(start_date='-1y', end_date='today'),  # Date within the last year
        'Interaction Outcome': interaction_outcome,
        'Notes': fake.text(max_nb_chars=200)  # Random notes about the interaction
    }
    crm_interactions.append(interaction)

# Convert to DataFrame
crm_df = pd.DataFrame(crm_interactions)

# Save the DataFrame to a CSV file
crm_df.to_csv('crm_interactions.csv', index=False)

# Display the first few rows of the DataFrame for verification
print(crm_df.head())
```
